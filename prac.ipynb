```python

import seaborn as sns


# Total votes per party
party_votes = mh.groupby('party')[['evm_votes', 'postal_votes']].sum()
party_votes['total_votes'] = party_votes['evm_votes'] + party_votes['postal_votes']
party_votes_sorted = party_votes.sort_values(by='total_votes', ascending=False)

# Plot 1: Total votes by party
plt.figure(figsize=(10, 6))
sns.barplot(x=party_votes_sorted.index, y=party_votes_sorted['total_votes'], palette='viridis')
plt.title('Total Votes by Party', fontsize=16)
plt.xlabel('Party', fontsize=12)
plt.ylabel('Total Votes', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Top 10 parties by total votes
constituency_votes = mh.groupby('constituency')[['evm_votes', 'postal_votes']].sum()
constituency_votes['total_votes'] = constituency_votes['evm_votes'] + constituency_votes['postal_votes']
constituency_votes_sorted = constituency_votes.sort_values(by='total_votes', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=constituency_votes_sorted.head(10).index, y=constituency_votes_sorted['total_votes'].head(10), palette='coolwarm')
plt.title('Top 10 Constituencies by Total Votes', fontsize=16)
plt.xlabel('Constituency', fontsize=12)
plt.ylabel('Total Votes', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Plot 3: Distribution of Total Votes (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(mh['total_votes'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Total Votes Per Candidate', fontsize=16)
plt.xlabel('Total Votes', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()

```

```python
# Identify the candidate with the highest votes
top_candidate = mh[['candidate', 'party', 'total_votes']].sort_values(by='total_votes', ascending=False).iloc[0]

# Print top candidate information
print("Top Candidate Information:")
print(f"Name: {top_candidate['candidate']}")
print(f"Party: {top_candidate['party']}")
print(f"Total Votes: {top_candidate['total_votes']}")
```
